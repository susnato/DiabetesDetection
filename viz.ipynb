```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./data/diabetes.csv")#.shape[0] * 0.8
```

```python
_ = data.Outcome.hist()
plt.title("Total data label dist")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data, train_size=0.8, stratify=data["Outcome"], random_state=42)
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))

train_data["Outcome"].hist(ax=ax[0]), test_data["Outcome"].hist(ax=ax[1])
ax[0].set_title("Train data label dist")
ax[1].set_title("Test data label dist")
plt.show()
```
